```python
#Practice the use of matplotlib
import numpy as np
import matplotlib.pyplot as plt
#Create a data set 
x = np.linspace(0, 10, 100)
#Print the data
print(x)
#Create a figure using plt.figure
fig = plt.figure()
#Plot the sine wave of x
plt.plot(x, np.sin(x), '-')
#Plot the cosine wave of x
plt.plot(x, np.cos(x), '.')
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Create a data set 
x = np.linspace(0, 10, 101)
#Print the data
print(x)
```

```python
# Create two arrays of 5 arbitrary numbers 
example1 = [1,3,5,2,9]
example3 = [5,2,6,4,9]

example2 = [2,4,6,7,8]
example4 = [5,1,9,3,2]
# Plot two bar charts using plt.bar and the arrays you've created above
# Label them "Example one" and "example two"
plt.bar(example1,example3, label="Example one", color='r')
plt.bar(example2,example4, label="Example two", color='g')
plt.legend()
# Label x axis "Bar number"
plt.xlabel("Bar number")
# Label y axis "Bar height"
plt.ylabel("Bar height")
# Set title "Bar Graph"
plt.title("Bar Graph")
# Show the chart
plt.show()
```

```python
# Histogram Code 
#Create an array of a random normal distribution
rannorm = np.random.laplace(loc=15, scale=3, size=100)
#Create a histogram using plt hist()
n, bins, patches = plt.hist(x=rannorm, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
#Set grid for x and y
#set transparency of the grid using alpha=0.5
plt.grid(axis='y', alpha = 0.5)
plt.grid(axis='x', alpha = 0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My First Histogram Ever')
plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
#Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```

```python
#Bonus: Create a scatter plot of random data
num_points = 50
x = np.random.rand(num_points)  #Random x values between 0 and 1
y = np.random.rand(num_points)  #Random y values between 0 and 1

#Create a scatter plot
plt.scatter(x, y, color='red', marker='o', label='Random Data')

#Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

#Add a legend
plt.legend()

#Show the plot
plt.show()

```
