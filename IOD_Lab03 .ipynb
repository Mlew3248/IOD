```python
#Load data from 'https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv'
#and print it
#First import libraries numpy, and pandas and matplotlib.pyplot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Load data from the above url using pandas read_csv
url = 'https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv'
weather = pd.read_csv(url)
#Print the top 10 rows
print(weather.head(10))
```

```python
# Use pandas describe() function to describe the data loaded above
weather.describe()

```

```python
#Plot temperature min and max
df = pd.DataFrame(weather)
minimum = df['Tmin']
maximum = df['Tmax']

#Create a line plot of minimum and maximum temperatures
plt.plot(maximum, label='Max Temperature')
plt.plot(minimum, label='Min Temperature')

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Temperature Variation in 2018 (London)')
plt.legend()

plt.show()
```

```python
#Plot the rain as above
rain = df['Rain']


#Create a line plot of minimum and maximum temperatures
plt.plot(rain, label='Rain')

plt.xlabel('Day')
plt.ylabel('Rain')
plt.title('Rain Variation in 2018 (London)')
plt.legend()

plt.show()
```

```python
#Plot the sun column

Sun = df['Sun']


#Create a line plot of minimum and maximum temperatures
plt.plot(Sun, label='Sun')

plt.xlabel('Day')
plt.ylabel('Sun')
plt.title('Sun Variation in 2018 (London)')
plt.legend()

plt.show()
```

```python
#Load the iris dataset
#Import numpy and pandas
import numpy as np
import pandas as pd
#Import seaborn library
import seaborn as sns
#Import load_iris
from sklearn.datasets import load_iris
#Load the dataset
dataset=load_iris()
#Place in pandas dataframe
data=pd.DataFrame(dataset["data"],columns=["Petal length","Petal Width","Sepal Length","Sepal Width"])
data["Species"]=dataset["target"]
data["Species"]=data["Species"].apply(lambda x: dataset["target_names"][x])
#Print top 10 rows
print(data.head(10))
```

```python
#Describe the dataframe
data.describe()
```

```python
#use seaborn pairplot function to plot relationship between variables

sns.pairplot(data, hue="Species")

plt.show()
```
